```python
from datetime import datetime

def smart_home(people):
    hour = datetime.now().hour
    now = datetime.now()

    if people == 0:
        print(f"Lights OFF (No one at home) time is={now.hour}:{now.minute:02d}")

    if hour >= 23:
        print(f"Lights OFF (Past 11:00 PM) time is={now.hour}:{now.minute:02d}")

    elif hour == 6:
        print(f"Lights ON in DIM mode (6:00 AM) time is={now.hour}:{now.minute:02d}")

    elif 6 <= hour < 18:
        print(f"Lights ON in DIM mode (Daytime)time is={now.hour}:{now.minute:02d}")

    else:
        print(f"Lights ON in BRIGHT mode (Nighttime) time is={now.hour}:{now.minute:02d}")

ppl = int(input("Enter number of people in the house: "))
smart_home(ppl)
```
