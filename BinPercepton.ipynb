```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
import random
```

```python
class Perceptron:
    def init(self):
        self.w_a = None
        self.w_b = None
        self.b = None

    def model(self, x_a, x_b):
        z = np.sum(self.w_a * x_a + self.w_b * x_b) + self.b
        return 1 if z >= 0 else 0 #activation function

    def predict(self, X_a ,X_b):
        Y=[]
        for a,b in zip(X_a,X_b):
            ans = self.model(a,b)
            Y.append(ans)
        return np.array(Y)

    def fit(self , X_a , X_b , Y , epochs = 1, lr = 1):
        self.w_a = np.ones(X_a.shape[0])
        self.w_b = np.ones(X_b.shape[0])
        self.b = 0

        accuracy = {}
        max_accu = 0
        wt_matrix = []

        for i in range(epochs):
            for x_a , x_b , y in zip(X_a,X_b , Y):
                y_pred = self.model(x_a,x_b)
                if y == 1 and y_pred == 0:
                    self.w_a = self.w_a + lr*x_a
                    self.w_b = self.w_b + lr*x_b
                    self.b = self.b - lr*1
                if y==0 and y_pred == 1:
                    self.w_a = self.w_a - lr*x_a
                    self.w_b = self.w_b - lr*x_b
                    self.b = self.b + lr *1
            wt_matrix.append((self.w_a.copy(), self.w_b.copy(), self.b))
            accuracy[i] = accuracy_score(self.predict(X_a,X_b),Y)
            if accuracy[i] > max_accu:
                max_accu = accuracy[i]
                s_w_a = self.w_a
                s_w_b = self.w_b
                s_b = self.b
        self.w_a = s_w_a
        self.w_b = s_w_b
        self.b = s_b
        print(f"the maximum accuracy is {max_accu}")
        return np.array(wt_matrix, dtype=object)


```

```python
X_a = np.array([[0],[0],[1],[1]])
X_b = np.array([[0],[1],[0],[1]])
Y_or = np.array([0,1,1,1])
Y_and = np.array([0,0,0,1])
```

```python
p = Perceptron()
p.fit(X_a,X_b,Y_or,epochs=10,lr=0.5)

print("Predictions for or:")
print(p.predict(X_a,X_b))
```
